```
jokes=[]

for i in range (1,101):
    jokes=jokes+[str(i)]
```

```
users=b   #length=2498300
movies=jokes*24983   #length=2498300 

ratings_dict={'userID':users,
              'itemID':movies,
              'rating':ratings}

df=pd.DataFrame(ratings_dict)
```

```
#remove this
newdf=df.iloc[0:1800005]
```

```
#write df in place of newdf for final run
reader=Reader(rating_scale=(-10,10))
trainset=Dataset.load_from_df(df[['userID','itemID','rating']],reader).build_full_trainset()
```

```
#write df in place of newdf for final run
data=Dataset.load_from_df(df[['userID','itemID','rating']],reader)
```

```
algo_svd=SVD()
algo_svd.fit(trainset)
```

```
cross_validate(algo_svd,data,measures=['rmse','mae'],cv=5,verbose=True)
```

```
algo_svd.predict(1,'2')
```

```
from surprise import KNNBasic
algo_knn=KNNBasic()
algo_knn.fit(trainset)
```

```
cross_validate(algo_knn,data,measures=['rmse','mae'],cv=5,verbose=True)
```

```
algo_knn.predict(43,'10')
```

```
from surprise import SVDpp
```

```
algo_SVDpp=SVDpp()
algo_SVDpp.fit(trainset)
```

```
cross_validate(algo_SVDpp,data,measures=['rmse','mae'],cv=2,verbose=True)
```
